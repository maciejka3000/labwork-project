```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
```

```python
data_folder = os.path.join(os.getcwd(), 'training_history')
files = glob.glob(os.path.join(data_folder, '*.csv'))

times = pd.DataFrame()
epoch_losses = pd.DataFrame()
val_losses = pd.DataFrame()

for i, file in enumerate(files):
    name_col = os.path.basename(file)
    read = pd.read_csv(file)
    print(read)
    time_col = read['time']
    times[name_col] = time_col

    epoch_col = read['loss']
    epoch_losses[name_col] = epoch_col

    val_col = read['val_loss']
    val_losses[name_col] = val_col

print(times)
times.plot()
epoch_losses.plot()
val_losses.plot()



```
